```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv")
data.head()
```

### Давайте посмотрим на все отделы, которые охвачены в этом бюджете:

```python
data
```

### Я вижу значение NaN в этом наборе данных, давайте удалим значения NaN и продолжим задачу анализа финансового бюджета с помощью Python:

```python
data.dropna()
```

### Я вижу, что не все отделы, включенные в этот набор данных, являются основными, поскольку некоторые отделы могут быть включены в другую категорию. Итак, давайте подготовим данные, выбрав только основные отделы и поместив все остальные отделы в другую категорию:

```python
data = data.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
data = data.append(row, ignore_index = True)
data
```

### Теперь давайте построим график этих данных, чтобы взглянуть на приоритеты правительства на финансовый год:

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

### Мы видим, что финансовый департамент получает большую часть общего бюджета правительства. Теперь давайте построим эти данные в виде кольцевой диаграммы, чтобы иметь четкое представление о распределении средств между всеми отделами:

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
